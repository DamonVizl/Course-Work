# Machine Learning Models



## Decision Tree

Take a data set and the machine assigns random values < or > of the condition and classify it.  
It is effectively a series of if/else statements.  
Once we have seperated (most of) the data into groups by doing this we have found a decision tree. We can feed new data into this tree to determine it's classification.  
The reason we need a machine to do this is it iterates through multiple splitting conditions to find the best set of splitting conditions/threshold values.  
  
To find the best splits it uses Information Gain. Basically which splits give us the most bang for our buck.  
The formula for the Information Gain is $IG = Entropy_p - \sum w_i*Entropy_c$  
Entropy is a measure of how much variance there is in the set. So if it is 50/50 the Entropy is 1 and if it is all of one classification (i.e. a pure leaf node) then the Entropy is 0. we want the entropy to be 0 but this won't always be the case. 
For binary classification if the entropy isnt 0 in the leaf node we can take the majority rule. If it is regression we can take the average. 





## Random Forrest

Multiple decision trees. 
The random nature is taken from the fact that we select a random assortment of data entries and a random selection of features, we train on these new randomely selected data sets and create a number (traditionally ~100) of decision trees. When new data comes in we can feed them into the forrest (all of the random trees) and then if it is a classification model we can take the most popular result (if it's binary and 70% of the results are True then we take the new point to be true) if it is a regression model then we take the mean of all the results.
